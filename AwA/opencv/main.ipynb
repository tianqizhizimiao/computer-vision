```python
import cv2
```

```python
# 读取图片
img = cv2.imread('image.png')  # 返回一个numpy图片数组 即图片像素数据
```

```python
# 保存图片
cv2.imwrite('save/image.png', img)
```

```python
# 缩放
resize_img = cv2.resize(img, (16, 16))  # 缩小为 16 x 16
cv2.imwrite('save/resize_image.png', resize_img)
```

```python
# 旋转
M = cv2.getRotationMatrix2D(
    [16, 16], # 旋转中心
    90,       # 旋转 90 度
    1.0       # 缩放比例 1.0 (不变)
)

rotate = cv2.warpAffine(
    img,      # img 对象
    M,        # 旋转变换对象
    (32, 32)  # 旋转后图片大小
)
cv2.imwrite('save/rotate_image.png', rotate)
```

```python
# 翻转
img = cv2.flip(
    img,  # 图片对象
    0     # 翻转操作 1为水平翻转    0为垂直翻转     -1 为水平垂直同时翻转
)

cv2.imwrite('save/flip_image.png', img)
```

```python
# 获取亮度与模糊度

import numpy as np

# 获取灰度图
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 获取亮度
print('亮度: ', img_gray.mean())  # 取平均值 0 ~ 255 之间 越高图越亮

# 获取模糊度
v = cv2.Laplacian(img_gray, cv2.CV_64F, ksize=1)
v = np.var(v)  # 求方差
print('模糊度: ', v)  # 越高越清晰
```
